# Conseguimos fazer filtros na base

```python
display(base)
```

```python
# Filtrando apenas programas que ficaram com retenção maior ou igual a 90%
base[base['Taxa de retenção'] >= 0.9]
```

```python
# Podemos fazer filtros usando E
base[(base['Taxa de retenção'] >= 0.9) & (base['Segundos'] > 20)]
```

```python
# Podemos fazer filtros usando OU
base[(base['Taxa de retenção'] >= 0.9) | (base['Custo'] < 50000)]
```

Uma forma muito útil de fazer seleção de dados é usando o *.loc()* ou o *.iloc()*

```python
# O .loc() permite fazer a busca da mesma forma acima
base.loc[base['Taxa de retenção'] >= 0.9]
```

```python
# Ele também permite utilizar argumentos lógicos
base.loc[(base['Taxa de retenção'] >= 0.9) | (base['Custo'] < 50000)]
```

```python
# Ele permite filtrar colunas de forma prática
base.loc[(base['Taxa de retenção'] >= 0.9) | (base['Custo'] < 50000), ['Venda $', 'Custo']]
```

Já o *iloc()* vai usar o índice para filtrar

```python
# Buscando as linhas de 7 (inclusive) a 13 (exclusive)
base.iloc[7:13]
```

```python
# Também posso usar para buscar apenas colunas específicas
base.iloc[7:13,[1, 3]]
```
